## Data Cleaning

### 1. Checking for Completeness

```python
# checking for missing values
nmes_df.isna().sum()
```

The Dataset seems to have no missing values, however , we know that the missing values are represented by '.' in the dataset. So we will replace '.' with NaN and then check for missing values.

```python
# replacing missing values with NaN initially represented by .
nmes_df = nmes_df.replace(".", np.NaN)
```

```python
nmes_df.sample(20)
```

```python
# checking for missing values again after replacing . with NaN
missing_values = nmes_df.isna().sum()
percent_missing = nmes_df.isna().sum()/nmes_df.shape[0]*100

value = {
    'missing_values ':missing_values,
    'percent_missing %':percent_missing
}
frame = pd.DataFrame(value)
frame
```

Based on the above analysis :
- The following columns have missing values above 50% : 

    - `current`  : 7211 missing values with 52%
    - `AGESMOKE` : 7407 missing values with 54%
    - `CIGSSMOK` : 7944	missing values with 58%
    - `CIGSADAY` : 10834 missing values with 79%
    - `AGESTOP`  : 10508 missing values with 77%
    

```python
# dropping columns with more than 50% missing values
nmes_df = nmes_df.dropna(thresh=0.5*len(nmes_df), axis=1)
```

```python
nmes_df.shape
```

```python
nmes_df.isna().sum()
```

```python
nmes_df.info()
```

```python
# Replacing the missing values with a describe word "missing" for the categorical variables
nmes_df['eversmk'].fillna('missing', inplace=True)
nmes_df['former'].fillna('missing', inplace=True)
nmes_df['beltuse'].fillna('missing', inplace=True)
nmes_df['marital'].fillna('missing', inplace=True)
nmes_df['POVSTALB'].fillna('missing', inplace=True)
```

```python
# Replacing the missing values with the mean for the numerical variables
nmes_df['packyears'] = nmes_df['packyears'].astype(float)
nmes_df['packyears'].fillna(nmes_df['packyears'].mean(), inplace=True)

nmes_df['yearsince'] = nmes_df['yearsince'].astype(float)
nmes_df['yearsince'].fillna(nmes_df['yearsince'].mean(), inplace=True)
```

```python
# confirming that there are no missing values
nmes_df.isna().sum()
```

```python
# changing the MALE column name to gender 
nmes_df = nmes_df.rename(columns={'MALE': 'gender'})
nmes_df = nmes_df.rename(columns={'LASTAGE': 'age'})
nmes_df.info()
```

```python
# converting the column from int to category
nmes_df['gender'] = nmes_df['gender'].astype('category')
nmes_df['RACE3'] = nmes_df['RACE3'].astype('category')
```

```python
# converting all the object data types to category
def convert_to_category(df):
    for column in df.columns:
        if df[column].dtype == 'object':
            df[column] = df[column].astype('category')
    return df

nmes_df = convert_to_category(nmes_df)
```
