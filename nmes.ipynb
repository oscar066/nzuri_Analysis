## Data Cleaning

### 1. Checking for Completeness

```python
# checking for missing values
nmes_df.isna().sum()
```

The Dataset seems to have no missing values, however , we know that the missing values are represented by '.' in the dataset. So we will replace '.' with NaN and then check for missing values.

```python
nmes_df.head(10)
```

```python
nmes_df.replace(r'\.', np.nan, regex=True, inplace=True)
nmes_df.to_csv('nmes3.csv', index=False)
nmes_df.head(10)
```

```python
nmes_df.sample(20)
```

```python
# checking for missing values again after replacing . with NaN
missing_values = nmes_df.isna().sum()
percent_missing = nmes_df.isna().sum()/nmes_df.shape[0]*100

value = {
    'missing_values ':missing_values,
    'percent_missing %':percent_missing
}
frame = pd.DataFrame(value)
frame
```

Based on the above analysis :
- The following columns have missing values above 50% : 

    - `current`  : 7211 missing values with 52%
    - `AGESMOKE` : 7407 missing values with 54%
    - `CIGSSMOK` : 7944	missing values with 58%
    - `CIGSADAY` : 10834 missing values with 79%
    - `AGESTOP`  : 10508 missing values with 77%
    

```python
# dropping columns with more than 50% missing values
nmes_df = nmes_df.dropna(thresh=0.5*len(nmes_df), axis=1)
```

```python
# Replacing the missing values with a describe word "missing" for the categorical variables
nmes_df['eversmk '].fillna('missing', inplace=True)
nmes_df['former '].fillna('missing', inplace=True)
nmes_df['beltuse '].fillna('missing', inplace=True)
nmes_df['marital '].fillna('missing', inplace=True)
nmes_df['POVSTALB'].fillna('missing', inplace=True)
```

```python
# Replacing the missing values with the mean for the numerical variables
nmes_df['packyears '] = nmes_df['packyears '].astype(float)
nmes_df['packyears '].fillna(nmes_df['packyears '].mean(), inplace=True)

nmes_df['yearsince '] = nmes_df['yearsince '].astype(float)
nmes_df['yearsince '].fillna(nmes_df['yearsince '].mean(), inplace=True)
```

## Model Building

```python
# baseline model
from sklearn.linear_model import LogisticRegression
log_reg = LogisticRegression()
log_reg.fit(X_train, y_train)

# making predictions
y_pred = log_reg.predict(X_test)

# evaluating the model
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
print("Accuracy score:", accuracy_score(y_test, y_pred))
print("Confusion matrix:\n", confusion_matrix(y_test, y_pred))
print("Classification report:\n", classification_report(y_test, y_pred))

# using cross validation
from sklearn.model_selection import cross_val_score
cv_score = cross_val_score(log_reg, X_train, y_train, cv=5)
print("CV scores:", cv_score)
print("CV mean:", cv_score.mean())
```

```python
# plot the ROC curve and compute AUC
from sklearn.metrics import roc_curve, roc_auc_score
y_pred_prob = log_reg.predict_proba(X_test)[:, 1]
fpr, tpr, thresholds = roc_curve(y_test, y_pred_prob)
plt.plot([0, 1], [0, 1], 'k--')
plt.plot(fpr, tpr, label='Logistic Regression')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Logistic Regression ROC Curve')
plt.show()
```

```python
# compute AUC
roc_auc_score(y_test, y_pred_prob)
```

```python
# using grid search
from sklearn.model_selection import GridSearchCV
params = {
    'C': [0.001, 0.01, 0.1, 1, 10, 100]
    }

grid_search = GridSearchCV(log_reg, params, cv=5)
grid_search.fit(X_train, y_train)

print("Best parameters:", grid_search.best_params_)
print("Best score:", grid_search.best_score_)
print("Best estimator:", grid_search.best_estimator_)
print("Best index:", grid_search.best_index_)
```

```python
# using the best estimator
log_reg = grid_search.best_estimator_
log_reg.fit(X_train, y_train)
y_pred = log_reg.predict(X_test)

# evaluating the model
print("Accuracy score:", accuracy_score(y_test, y_pred))
print("Confusion matrix:\n", confusion_matrix(y_test, y_pred))
print("Classification report:\n", classification_report(y_test, y_pred))

# using cross validation
cv_score = cross_val_score(log_reg, X_train, y_train, cv=5)
print("CV scores:", cv_score)
print("CV mean:", cv_score.mean())
```

```python
# Decision tree classifier
from sklearn.tree import DecisionTreeClassifier
tree_clf = DecisionTreeClassifier()
tree_clf.fit(X_train, y_train)

# making predictions
y_pred = tree_clf.predict(X_test)

# evaluating the model
print("Accuracy score:", accuracy_score(y_test, y_pred))
print("Confusion matrix:\n", confusion_matrix(y_test, y_pred))
print("Classification report:\n", classification_report(y_test, y_pred))

# using cross validation
cv_score = cross_val_score(tree_clf, X_train, y_train, cv=5)
print("CV scores:", cv_score)
print("CV mean:", cv_score.mean())
```

```python
# plot ROC curve and calculate AUC
from sklearn.metrics import roc_curve, roc_auc_score
y_pred_proba = tree_clf.predict_proba(X_test)[:, 1]
fpr, tpr, thresholds = roc_curve(y_test, y_pred_proba)
plt.plot([0, 1], [0, 1], 'k--')
plt.plot(fpr, tpr, label="Decision Tree")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("Decision Tree ROC Curve")
plt.show()
```

```python
# calculate AUC
roc_auc_score(y_test, y_pred_proba)
```

```python
# using grid search
params = {
    'max_depth': [2, 4, 6, 8, 10]
    }

grid_search = GridSearchCV(tree_clf, params, cv=5)
grid_search.fit(X_train, y_train)

print("Best parameters:", grid_search.best_params_)
print("Best score:", grid_search.best_score_)
print("Best estimator:", grid_search.best_estimator_)
print("Best index:", grid_search.best_index_)

tree_clf = grid_search.best_estimator_
```

```python
# using the best estimator
tree_clf.fit(X_train, y_train)
y_pred = tree_clf.predict(X_test)

# evaluating the model
print("Accuracy score:", accuracy_score(y_test, y_pred))
print("Confusion matrix:\n", confusion_matrix(y_test, y_pred))
print("Classification report:\n", classification_report(y_test, y_pred))

# using cross validation
cv_score = cross_val_score(tree_clf, X_train, y_train, cv=5)
print("CV scores:", cv_score)
print("CV mean:", cv_score.mean())
```

```python
# Random forest classifier
from sklearn.ensemble import RandomForestClassifier
forest_clf = RandomForestClassifier()
forest_clf.fit(X_train, y_train)

# making predictions
y_pred = forest_clf.predict(X_test)

# evaluating the model
print("Accuracy score:", accuracy_score(y_test, y_pred))
print("Confusion matrix:\n", confusion_matrix(y_test, y_pred))
print("Classification report:\n", classification_report(y_test, y_pred))

# using cross validation
cv_score = cross_val_score(forest_clf, X_train, y_train, cv=5)
print("CV scores:", cv_score)
print("CV mean:", cv_score.mean())
```

```python
# plot ROC curve and calculate AUC
y_pred_proba = forest_clf.predict_proba(X_test)[:, 1]
fpr, tpr, thresholds = roc_curve(y_test, y_pred_proba)
plt.plot([0, 1], [0, 1], 'k--')
plt.plot(fpr, tpr, label="Random Forest")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("Random Forest ROC Curve")
plt.show()
```

```python
# calculate AUC
roc_auc_score(y_test, y_pred_proba)
```

```python
# using grid search
params = {
    'n_estimators': [100, 200, 300, 400, 500]
    }
grid_search = GridSearchCV(forest_clf, params, cv=5)
grid_search.fit(X_train, y_train)

print("Best parameters:", grid_search.best_params_)
print("Best score:", grid_search.best_score_)
print("Best estimator:", grid_search.best_estimator_)
print("Best index:", grid_search.best_index_)

forest_clf = grid_search.best_estimator_
```

```python
# using the best estimator
forest_clf.fit(X_train, y_train)
y_pred = forest_clf.predict(X_test)

# evaluating the model
print("Accuracy score:", accuracy_score(y_test, y_pred))
print("Confusion matrix:\n", confusion_matrix(y_test, y_pred))
print("Classification report:\n", classification_report(y_test, y_pred))

# using cross validation
cv_score = cross_val_score(forest_clf, X_train, y_train, cv=5)
print("CV scores:", cv_score)
print("CV mean:", cv_score.mean())
```
