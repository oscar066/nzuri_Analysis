## Data Cleaning

### 1. Checking for Completeness

```python
# checking for missing values
nmes_df.isna().sum()
```

The Dataset seems to have no missing values, however , we know that the missing values are represented by '.' in the dataset. So we will replace '.' with NaN and then check for missing values.

```python
nmes_df.head(10)
```

```python
# Relacing '.' which represents missing values with NaN
nmes_df.replace(r'\.', np.nan, regex=True, inplace=True)
nmes_df.to_csv('nmes3.csv', index=False)
nmes_df.head(10)
```

```python
# checking for missing values again after replacing . with NaN
missing_values = nmes_df.isna().sum()
percent_missing = nmes_df.isna().sum()/nmes_df.shape[0]*100

value = {
    'missing_values ':missing_values,
    'percent_missing %':percent_missing
}
frame = pd.DataFrame(value)
frame
```

Based on the above analysis :
- The following columns have missing values above 50% : 

    - `current`  : 7211 missing values with 52%
    - `AGESMOKE` : 7407 missing values with 54%
    - `CIGSSMOK` : 7944	missing values with 58%
    - `CIGSADAY` : 10834 missing values with 79%
    - `AGESTOP`  : 10508 missing values with 77%
    

```python
# dropping columns with more than 50% missing values
nmes_df = nmes_df.dropna(thresh=0.5*len(nmes_df), axis=1)
```

## Data Preparation
- encoding categorical variables
- scaling numerical variables
- dealing with class imbalance

#### 1. Encoding Categorical Variables

```python
# encoding the categorical variables
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()
for column in nmes_df.columns:
    if nmes_df[column].dtype == 'category':
        nmes_df[column] = label_encoder.fit_transform(nmes_df[column])
```

#### 2. Scaling Numerical Variables

```python
# scaling the data
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
nmes_df_scaled = scaler.fit_transform(nmes_df)

# creating a dataframe from the scaled data
nmes_df_scaled = pd.DataFrame(nmes_df_scaled, columns=nmes_df.columns)
nmes_df_scaled.head()
```

#### 3. Feature Selection

```python
# feature selection
from sklearn.feature_selection import SelectKBest, chi2

X = nmes_df.drop('target', axis=1)
y = nmes_df['target']

# apply SelectKBest class to extract top 10 best features
bestfeatures = SelectKBest(score_func=chi2, k=10)
fit = bestfeatures.fit(X, y)
dfscores = pd.DataFrame(fit.scores_)
dfcolumns = pd.DataFrame(X.columns)

# concat two dataframes for better visualization
featureScores = pd.concat([dfcolumns, dfscores], axis=1)
featureScores.columns = ['Specs', 'Score']  # naming the dataframe columns
print(featureScores.nlargest(10, 'Score'))  # print 10 best features
```

#### 4. Dealing with Class Imbalance

```python
# handling class imbalance 
from imblearn.over_sampling import SMOTE

smote = SMOTE(random_state=42, sampling_strategy='minority')
X_smote, y_smote = smote.fit_resample(nmes_df_scaled, nmes_df['target'])

# splitting the data into train and test sets
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_smote, y_smote, test_size=0.25, random_state=42)
```

## Model Building

#### 1. Logistic Regression

```python
# baseline model
from sklearn.linear_model import LogisticRegression
log_reg = LogisticRegression()
log_reg.fit(X_train, y_train)

# making predictions
y_pred = log_reg.predict(X_test)

# evaluating the model
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
print("Accuracy score:", accuracy_score(y_test, y_pred))
print("Confusion matrix:\n", confusion_matrix(y_test, y_pred))
print("Classification report:\n", classification_report(y_test, y_pred))

# using cross validation
from sklearn.model_selection import cross_val_score
cv_score = cross_val_score(log_reg, X_train, y_train, cv=5)
print("CV scores:", cv_score)
print("CV mean:", cv_score.mean())
```

```python
# plot the ROC curve and compute AUC
from sklearn.metrics import roc_curve, roc_auc_score
y_pred_prob = log_reg.predict_proba(X_test)[:, 1]
fpr, tpr, thresholds = roc_curve(y_test, y_pred_prob)
plt.plot([0, 1], [0, 1], 'k--')
plt.plot(fpr, tpr, label='Logistic Regression')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Logistic Regression ROC Curve')
plt.show()
```

```python
# compute AUC
roc_auc_score(y_test, y_pred_prob)
```

#### 2. Decision tree

```python
# Decision tree classifier
from sklearn.tree import DecisionTreeClassifier
tree_clf = DecisionTreeClassifier()
tree_clf.fit(X_train, y_train)

# making predictions
y_pred = tree_clf.predict(X_test)

# evaluating the model
print("Accuracy score:", accuracy_score(y_test, y_pred))
print("Confusion matrix:\n", confusion_matrix(y_test, y_pred))
print("Classification report:\n", classification_report(y_test, y_pred))

# using cross validation
cv_score = cross_val_score(tree_clf, X_train, y_train, cv=5)
print("CV scores:", cv_score)
print("CV mean:", cv_score.mean())
```

```python
# plot ROC curve and calculate AUC
from sklearn.metrics import roc_curve, roc_auc_score
y_pred_proba = tree_clf.predict_proba(X_test)[:, 1]
fpr, tpr, thresholds = roc_curve(y_test, y_pred_proba)
plt.plot([0, 1], [0, 1], 'k--')
plt.plot(fpr, tpr, label="Decision Tree")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("Decision Tree ROC Curve")
plt.show()
```

```python
# calculate AUC
roc_auc_score(y_test, y_pred_proba)
```

#### 3. Random Forest

```python
# Random forest classifier
from sklearn.ensemble import RandomForestClassifier
forest_clf = RandomForestClassifier()
forest_clf.fit(X_train, y_train)

# making predictions
y_pred = forest_clf.predict(X_test)

# evaluating the model
print("Accuracy score:", accuracy_score(y_test, y_pred))
print("Confusion matrix:\n", confusion_matrix(y_test, y_pred))
print("Classification report:\n", classification_report(y_test, y_pred))

# using cross validation
cv_score = cross_val_score(forest_clf, X_train, y_train, cv=5)
print("CV scores:", cv_score)
print("CV mean:", cv_score.mean())
```

```python
# plot ROC curve and calculate AUC
y_pred_proba = forest_clf.predict_proba(X_test)[:, 1]
fpr, tpr, thresholds = roc_curve(y_test, y_pred_proba)
plt.plot([0, 1], [0, 1], 'k--')
plt.plot(fpr, tpr, label="Random Forest")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("Random Forest ROC Curve")
plt.show()
```

```python
# calculate AUC
roc_auc_score(y_test, y_pred_proba)
```
